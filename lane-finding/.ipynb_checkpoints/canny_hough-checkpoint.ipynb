```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
%matplotlib inline
```

```python
image = mpimg.imread('exit-ramp.png')
plt.imshow(image)
```

```python
import cv2
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
low_threshold = 50
high_threshold = 150

edges = cv2.Canny(np.uint8(gray * 255), low_threshold, high_threshold)
plt.imshow(edges, cmap='gray')
```

```python
# 4-sided polygon mask

mask = np.zeros_like(edges)   
ignore_mask_color = 255   
vertices = np.array([[(80,539),(420, 300), (520, 300), (890,539)]], dtype=np.int32)
cv2.fillPoly(mask, vertices, ignore_mask_color)
masked_edges = cv2.bitwise_and(edges, mask)

plt.imshow(masked_edges, cmap='gray')
```

```python
rho = 1
theta = np.pi/180
threshold = 50
min_line_length = 10
max_line_gap = 3
line_image = np.uint8(np.copy(image)*0) # creating a blank to draw lines on
lines = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]),
                            min_line_length, max_line_gap)

for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0), 10)

color_edges = np.dstack((edges, edges, edges))
line_edges = cv2.addWeighted(line_image, .8, color_edges, 1, 0)
plt.imshow(line_edges)
```
